```python
import pandas as pd
```

```python
df1 = pd.read_csv('Bevölkerung.csv')

print(len(df1))
df1.head()
```

```python
df1.info()
```

```python
# Clean out not useful data:
df1 = df1[['StichtagDatJahr', 'AlterV20Kurz', 'KreisLang', 'QuarLang', 'AnzBestWir']]
df1
```

> We are only interested in the latest data (2022):

```python
df1 = df1[df1['StichtagDatJahr'] == 2022].reset_index(drop=True)
df1
```

> There is multiple mesurements:

```python
df1[(df1['AlterV20Kurz'] == '0-19') & (df1['QuarLang'] == 'City')].head()
```

> Just add them together:

```python
df1 = df1.groupby(['KreisLang', 'QuarLang','AlterV20Kurz']).AnzBestWir.count().reset_index()
df1.head()
```

```python
# They are not all the same:
df1[df1['AnzBestWir'] != 12]
```

```python
df1.info()
```

> Get the ID of the Stadtkreis:

```python
df1['id_Kreis'] = df1.KreisLang.apply(lambda x: x.split(' ')[1] if x != 'Unbekannt' else '-')
df1
```

```python
# Drop where no ID:
df1 = df1.drop(df1[df1['id_Kreis'] == '-'].index)
df1
```

```python
df1.to_csv('Modified_data/Bevölkerung.csv')
```
